# Questão 1

Descreva as funções básicas de uma UCP, indicando os seus componentes principais.

Resolução:

As funções básicas de uma UCP (Unidade Central de Processamento) consistem em atuar como o "cérebro" do computador, executando o ciclo de busca, decodificação e execução das instruções vindas da memória. Para desempenhar esse papel, ela depende de componentes essenciais interligados:
Unidade de Controlo (UC)
É a peça-chave que comanda todo o funcionamento da UCP. Recebe a instrução da memória, decodifica-a (traduzindo o que deve ser feito) e gera os sinais elétricos de temporização que coordenam as demais unidades no momento exato.
Unidade Lógica e Aritmética (ULA / ALU)
Responsável pela execução propriamente dita. É nela que acontecem os cálculos matemáticos (soma, subtração, multiplicação e divisão) e as comparações lógicas (maior/menor, igualdade, AND, OR, NOT e XOR).
Registadores internos
São pequenas memórias de altíssima velocidade localizadas dentro da UCP, usadas para guardar dados temporariamente durante o processamento. Os principais são:

RI (Registador de Instrução) — guarda a instrução atual que está sendo decodificada e executada.
PC (Contador de Programa) — guarda o endereço (posição na memória) da próxima instrução a ser buscada.
ACC (Acumulador) — armazena temporariamente o resultado imediato das operações feitas pela ULA, sendo frequentemente utilizado para acumular somas sucessivas.

Além destes, existem ainda os registadores de endereço (REM) e de dados (RDM), usados na troca de informação com a memória principal, e os barramentos internos (de dados, endereços e controlo), que interligam todos esses componentes entre si e com os demais dispositivos do computador.

# Questão 2

Quais são as funções da unidade aritmética e lógica - UAL?

Resolução:

A UAL (Unidade Aritmética e Lógica) pode ser comparada à "calculadora" da UCP, pois é responsável por executar as operações aritméticas e lógicas determinadas pela Unidade de Controle. Suas principais funções são realizar operações matemáticas, como soma, subtração, multiplicação e divisão (embora algumas arquiteturas mais simples não possuam instruções de multiplicação e divisão implementadas diretamente em hardware).

Além disso, a UAL realiza comparações entre valores, verificando condições como igualdade, maior que e menor que, e executa operações lógicas, como AND, OR, NOT e XOR.

Em algumas arquiteturas, a UAL também executa operações de deslocamento (shift) e rotação (rotate) de bits. Os deslocamentos podem ser utilizados para multiplicar ou dividir valores por potências de 2, enquanto as rotações reorganizam os bits de um operando, sendo úteis em determinadas operações de processamento e manipulação de dados.

Ao final de cada operação, a UAL atualiza as flags (bits de status), que indicam características do resultado, como Zero (Z), quando o resultado é zero; Negativo (N) ou Sinal (S), quando o resultado é negativo; e Carry (C), quando ocorre transporte de bit em operações aritméticas.

# Questão 3

O que é e para que servia o ACC?

Resolução:

O ACC (Acumulador) é um registrador interno da UCP utilizado para armazenar temporariamente os resultados das operações realizadas pela UAL. Ele serve como uma área de trabalho para a CPU, permitindo que resultados intermediários sejam guardados e reutilizados em operações seguintes, sem a necessidade de acessar a memória principal, o que torna o processamento mais rápido.

# Questão 5

Quais são as funções da unidade de controle de um processador?

Resolução:
A Unidade de Controle (UC) é responsável por coordenar todas as operações da UCP, atuando como o "maestro" do processador. Ela busca as instruções armazenadas na memória, decodifica seu significado e gera os sinais de controle que sincronizam a execução das operações pela UAL, pelos registradores e pelos demais componentes do sistema. Além disso, controla o fluxo de dados entre a CPU, a memória e os dispositivos de entrada e saída, garantindo que as instruções sejam executadas corretamente e na sequência adequada.

# Questão 7

Considere um computador cuja MP é organizada com N células de 1 byte cada uma. As instruções interpretadas pela UCP possuem três tamanhos diferentes: as do tipo A possuem 16 bits; as do tipo 13 têm 32 bits e as do tipo C possuem 48 bits. Considerando que o código de operação de cada uma tem um tamanho fixo e igual a 8 bits e que os programas executados nesse processador são constituídos de uma mistura dos três tipos de instruções, imagine um processo prático para incremento automático do CI após a execução de cada instrução de um programa.

Resolução:

Como as instruções possuem tamanhos diferentes (16, 32 e 48 bits), o incremento do CI não pode ser fixo. Após a busca da instrução, a Unidade de Controle identifica seu tamanho por meio do código de operação (opcode) de 8 bits durante a decodificação. Com base nisso, o CI é incrementado pelo número de bytes correspondente ao tamanho da instrução:

Tipo A: 16 bits = 2 bytes → CI = CI + 2;
Tipo B: 32 bits = 4 bytes → CI = CI + 4;
Tipo C: 48 bits = 6 bytes → CI = CI + 6.

Dessa forma, ao término da execução de cada instrução, o CI passa a apontar para o endereço inicial da próxima instrução, garantindo a execução correta do programa.

# Questão 8

Considere um processador cujo ciclo de instrução não possua a etapa de incremento automático do valor do CI. Imagine um método alternativo que permita a execução do programa.

Resolução:

Se o processador não possuísse a etapa de incremento automático do CI, seria necessário que o próprio programa atualizasse o valor do CI após a execução de cada instrução. Assim, ao final de cada instrução, uma instrução adicional calcularia o endereço da próxima instrução e gravaria esse valor no CI. Dessa forma, o processador continuaria executando o programa corretamente, porém com menor desempenho devido à necessidade de executar uma instrução extra para atualizar o CI.

Outra alternativa seria utilizar uma instrução de JUMP ao final de cada instrução, indicando explicitamente o endereço da próxima instrução a ser executada. Também seria possível projetar o conjunto de instruções de forma que cada instrução armazenasse, além do código de operação e dos operandos, o endereço da próxima instrução, eliminando a necessidade do incremento automático do CI. Essa abordagem, entretanto, aumentaria o tamanho das instruções e o consumo de memória.

# Questão 11

Qual é o registrador cujo conteúdo determina a capacidade de memória de um computador?

Resolução:

O REM (Registrador de Endereço de Memória) é o registrador cujo tamanho determina a capacidade máxima de memória que um computador pode endereçar. Se ele possui n bits, consegue representar 2ⁿ endereços diferentes. Assim, ele determina a capacidade máxima de memória endereçável pelo computador. Ele está localizado na UCP e é responsável por armazenar o endereço da posição de memória que será acessada, seja para leitura ou escrita. Quanto maior for o número de bits do REM, maior será a quantidade de endereços de memória que o processador poderá acessar.

# Questão 15

Quando se fala que um determinado microcomputador A é um micro de 8 bits e que um outro micro 13(acho que é D) é de 16 bits, a que estamos nos referindos? Ao tamanho da célula de MP ou ao tamanho da palavra? Qual a base desses dois conceitos (palavra e célula)?

Resolução:

Quando se diz que um microcomputador é de 8 bits ou de 16 bits, está se referindo ao tamanho da palavra do processador, e não ao tamanho da célula da memória principal. O tamanho da palavra corresponde à quantidade de bits que a UCP consegue processar em uma única operação, influenciando o tamanho dos registradores, da UAL e do barramento de dados.

Já a célula de memória é a menor unidade endereçável da memória principal, sendo normalmente composta por 1 byte (8 bits) nos computadores atuais.

Portanto, a palavra é definida pela arquitetura do processador e representa a quantidade de bits processados simultaneamente pela UCP, enquanto a célula de memória é definida pela organização da memória e representa a menor unidade de armazenamento que pode ser acessada por um endereço.

<a name="questao_18"></a>
# Questão 18

Explique a diferença entre um processamento sequêncial e um outro pipeline.

Resolução:

O processamento sequencial executa uma instrução de cada vez. Ou seja, a próxima instrução só começa a ser processada quando a anterior termina todas as suas etapas (busca, decodificação, execução, etc.). Esse método é mais simples, porém utiliza menos eficientemente os recursos do processador.

Já o processamento pipeline divide a execução das instruções em estágios. Enquanto uma instrução está sendo executada, outra pode estar sendo decodificada e uma terceira pode estar sendo buscada na memória. Assim, várias instruções são processadas simultaneamente em diferentes etapas, aumentando o desempenho e a taxa de execução, embora cada instrução individual continue passando pelos mesmos estágios.
